```r
df <- df[!(is.na(df$educ) | is.na(df$wrkstat)), ]
```

```r
df$emailtotal <- ifelse(is.na(df$emailhr) | is.na(df$emailmin), NA,
                        df$emailhr * 60 + df$emailmin)
```

```r
df <- df[, !(names(df) %in% c("emailhr", "emailmin"))]
```

```r
df
```

```r
df_non_null <- df[!is.na(df$emailtotal), ]
```

```r
dim(df_non_null)
```

```r
df_non_null <- replace(df_non_null, is.na(df_non_null), "Unknown")
```

```r
df_non_null
```

```r
df_non_null$educ <- factor(df_non_null$educ)
```

```r
categorical_columns <- c("harass5", "educ", "polviews", "advfront", "snapchat", "instagrm")

one_hot_encoded <- model.matrix(~ . - 1, data = df_non_null[, categorical_columns])
```

```r
 dim(one_hot_encoded)
```

```r
one_hot_encoded
```

```r
numerical_columns <- c('emailtotal')

# Scale the numerical columns using the scale function
df_non_null[numerical_columns] <- scale(df_non_null[numerical_columns])
```

```r
one_hot_encoded <- cbind(one_hot_encoded, df_non_null$emailtotal)
```

```r
one_hot_encoded <- cbind(one_hot_encoded, df_non_null$wrkstat)
```

```r
one_hot_encoded <- data.frame(one_hot_encoded)
```

```r
colnames(one_hot_encoded)[colnames(one_hot_encoded) == "df_non_null$wrkstat"] <- "wrkstat"
```

```r
one_hot_encoded$wrkstat <- as.factor(one_hot_encoded$wrkstat)
```

```r
fit_nnet <- multinom(wrkstat ~ ., one_hot_encoded)
```

```r
summary(fit_nnet)
```

```r
initial_model <- multinom(wrkstat ~ ., data = one_hot_encoded)
```

```r
selected_model <- stepAIC(initial_model, direction = "both", trace = FALSE)
```

```r
# Print the selected model
summary(selected_model)
```
